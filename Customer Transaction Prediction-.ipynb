<h4>Threshold Tuning on HistGradientBoosting(basemodel)</h4>

* tried with scaled data still got the same result ,so going with X_train and X_test

```python
from sklearn.metrics import precision_recall_curve
hgb_probs = hgb_model.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(Y_test, hgb_probs)
best_threshold = thresholds[np.argmax(
    2 * (precision * recall) / (precision + recall)
)]
hgb_pred_custom = (hgb_probs >= best_threshold).astype(int)
```

```python
print (classification_report(Y_test,hgb_pred_custom))
print(confusion_matrix(Y_test,hgb_pred_custom))
print("ROC_AUC:",roc_auc_score(Y_test, hgb_probs))
```

<h4>Threshold tuning on naive bayes</h4>

```python
nb_probs = nb.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(Y_test, nb_probs)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-8)
best_threshold_nb = thresholds[np.argmax(f1_scores)]
nb_pred_tuned = (nb_probs >= best_threshold_nb).astype(int)
```

```python
print("Naive Bayes Best Threshold:", best_threshold_nb)
print(confusion_matrix(Y_test, nb_pred_tuned))
print(classification_report(Y_test, nb_pred_tuned))
```

<h4>Automated Hperparameter Tuning</h4>

```python
from sklearn.model_selection import RandomizedSearchCV
param_dist = {
    'max_depth': [3, 4, 5, 6, 8],
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'max_iter': [100, 200, 300],
    'min_samples_leaf': [20, 50, 100]
}
hgb = HistGradientBoostingClassifier(random_state=42)
random_search = RandomizedSearchCV(
    hgb,
    param_distributions=param_dist,
    scoring='roc_auc',
    n_iter=20,
    cv=3,
    n_jobs=-1,
    random_state=42
)

random_search.fit(X_train, Y_train)
best_hgb = random_search.best_estimator_
```

<h4>Rerun threshold tuning</h4>

```python
from sklearn.metrics import precision_recall_curve
hgb_probs = best_hgb.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(Y_test, hgb_probs)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-8)
best_threshold = thresholds[np.argmax(f1_scores)]
hgb_pred_custom = (hgb_probs >= best_threshold).astype(int)
```

```python
print (classification_report(Y_test,hgb_pred_custom))
print(confusion_matrix(Y_test,hgb_pred_custom))
print("ROC_AUC:",roc_auc_score(Y_test, hgb_probs))
```

* increase in ROC_AUC and overall performance

<h4>Permutation Feature Importance</h4>

permutation feature imporatnace work by shuffling one feature at a time,if model performance drop ,feature is imporatant

```python
from sklearn.inspection import permutation_importance
perm_importance = permutation_importance(
    best_hgb,
    X_test,
    Y_test,
    scoring='roc_auc',
    n_repeats=10,
    random_state=42,
    n_jobs=-1
)
feature_importance = pd.DataFrame({
    'Feature': X_test.columns,
    'Importance': perm_importance.importances_mean
}).sort_values(by='Importance', ascending=False)
feature_importance.head(10)
```

* var_81 is the feature that is influencing the target the most
* shuffling var_81 affected the model performance the most

<h4>checking for Overfitting or Underfitting</h4>

<h4>Cross validation</h4>

```python
from sklearn.model_selection import cross_val_score

cv_auc = cross_val_score(
    best_hgb,
    X_train,
    Y_train,
    scoring='roc_auc',
    cv=3,
    n_jobs=-1
)

print("CV ROC-AUC mean:", cv_auc.mean())
print("CV ROC-AUC std:", cv_auc.std())
```
